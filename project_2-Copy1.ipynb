# The Three Body Problem

Trying to predict the motion of three bodies drove Isaac Netwon crazy until he gave up. It turns out that there is no real general analytical solution to this problem. Instead visualizing this problem we end up with chaotic solutions. The three body problem could be summarized with the following equations:

Solutions to the three body problem were found... but for very specific cases. Today we will explore the "figure 8" solution and the pythagorean three body problem solution. Both are stable but for small perturbations. We will also explore the stability of the solutions at the lagrangian points that depend on the mass.

Here is the code for the differntial equation code:

```python

def dSdt(x1, y1, x2, y2, x3, y3, vx1, vy1, vx2, vy2, vx3, vy3):
    
    r12 = np.sqrt((x2-x1)**2 + (y2-y1)**2)
    r13 = np.sqrt((x3-x1)**2 + (y3-y1)**2)
    r23 = np.sqrt((x2-x3)**2 + (y2-y3)**2)
    return [ vx1,
            vy1,
            vx2,
            vy2,
            vx3,
            vy3,
            m2/r12**3 * (x2-x1) + m3/r13**3 * (x3-x1), #mass 1
            m2/r12**3 * (y2-y1) + m3/r13**3 * (y3-y1),
            m1/r12**3 * (x1-x2) + m3/r23**3 * (x3-x2), #mass 2
            m1/r12**3 * (y1-y2) + m3/r23**3 * (y3-y2),
            m1/r13**3 * (x1-x3) + m2/r23**3 * (x2-x3), #mass 3
            m1/r13**3 * (y1-y3) + m2/r23**3 * (y2-y3)
           ]
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
```

```python
#euler method

def forwardEuler(f, x, t, dt, n):
    T = np.array(t)
    X = np.array(x)
    for i in range(n):
        t += dt
        x += np.array(f(*x)) * dt
        T = np.append( T, t )
        X = np.vstack((X, x))
    return T, X


#specific inital conditions for figure 8 shape

m1 = 1
m2 = 1 
m3 = 1
x1_0 = 0.9700436
y1_0 = -0.24308753
x2_0 = -0.9700436
y2_0 = 0.24308753
x3_0 = 0
y3_0 = 0
vx1_0 = 0.466203685
vy1_0 = 0.43236573
vx2_0 = 0.466203685
vy2_0 = 0.43236573
vx3_0 = -0.93240737
vy3_0 = -0.86473146


#######################

N=1000 # we will use 1000 steps

T, X = forwardEuler(dSdt,(
    x1_0,  y1_0,                    
    x2_0,  y2_0,
    x3_0,  y3_0,
    vx1_0, vy2_0,
    vx2_0, vy2_0,
    vx3_0, vy3_0
                           ), 0, 100/N, N)


#plotting the trajectory for the three planets

plt.plot(X[:,0], X[:,1])
plt.plot(X[:,2], X[:,3])
plt.plot(X[:,4], X[:,5])
plt.xlim(-2,2)
plt.ylim(-1,1)

```

As you notice the higher the number of steps the closer we get to an infinity shaped figure. This just demonstrates how chaotic this system is. Let us try the RK4 method instead:

```python
def RK4(f, x, t, dt, n):    #the RK4 method
    T = np.array(t)
    X = np.array(x)
    
    for i in range(n):
        k1 = dt * np.array(f(*(x         )))
        k2 = dt * np.array(f(*(x + 0.5*k1)))
        k3 = dt * np.array(f(*(x + 0.5*k2)))
        k4 = dt * np.array(f(*(x +     k3)))
        
        t += dt
        x += k1/6 + k2/3 + k3/3 + k4/6
        
        T = np.append( T, t)
        X = np.vstack((X, x))        
    return T, X


#specific inital conditions for figure 8 shape

m1 = 1
m2 = 1 
m3 = 1
x1_0 = 0.9700436
y1_0 = -0.24308753
x2_0 = -0.9700436
y2_0 = 0.24308753
x3_0 = 0
y3_0 = 0
vx1_0 = 0.466203685
vy1_0 = 0.43236573
vx2_0 = 0.466203685
vy2_0 = 0.43236573
vx3_0 = -0.93240737
vy3_0 = -0.86473146


#######################

N=1000   # we will use 1000 steps

T, X = RK4(dSdt,(
    x1_0, y1_0,
    x2_0, y2_0,
    x3_0, y3_0,
    vx1_0, vy2_0,
    vx2_0, vy2_0,
    vx3_0, vy3_0
                ), 0, 100/N, N)

plt.plot(X[:,0], X[:,1])
plt.plot(X[:,2], X[:,3])
plt.plot(X[:,4], X[:,5])
plt.xlim(-2,2)
plt.ylim(-1,1)
```

# Pythagorean Three Body Problem

The Pythagorean Body Problem is a specific case of the three body problem where the ratio of the masses is 3:4:5 and the initial distance between the bodies is the sides of 3,4,5 scalene right triangle so that the opposite side to each mass is its number in the ratio. The initial velocities of all the planets is zero.

$v{y1,0}$ = 0

We notice that even the RK4 method for a large number of steps doesn't display the stable nature of this solution. Let us use a very specific scipy differential equation solver with high accuracy instead:

```python
def dSdt1(t, S):
    x1, y1, x2, y2, x3, y3, vx1, vy1, vx2, vy2, vx3, vy3 = S
    r12 = np.sqrt((x2-x1)**2 + (y2-y1)**2)
    r13 = np.sqrt((x3-x1)**2 + (y3-y1)**2)
    r23 = np.sqrt((x2-x3)**2 + (y2-y3)**2)
    return [ vx1,
            vy1,
            vx2,
            vy2,
            vx3,
            vy3,
            m2/r12**3 * (x2-x1) + m3/r13**3 * (x3-x1), #mass 1
            m2/r12**3 * (y2-y1) + m3/r13**3 * (y3-y1),
            m1/r12**3 * (x1-x2) + m3/r23**3 * (x3-x2), #mass 2
            m1/r12**3 * (y1-y2) + m3/r23**3 * (y3-y2),
            m1/r13**3 * (x1-x3) + m2/r23**3 * (x2-x3), #mass 3
            m1/r13**3 * (y1-y3) + m2/r23**3 * (y2-y3)
           ]

t = np.linspace(0, 20, 1000)

sol = solve_ivp(dSdt1, (0,20), y0=[x1_0, y1_0, x2_0, y2_0, x3_0, y3_0,
                       vx1_0, vy1_0, vx2_0, vy2_0, vx3_0, vy3_0],
                method = 'DOP853', t_eval=t, rtol=1e-10, atol=1e-13)

t = sol.t
x1 = sol.y[0]
y1 = sol.y[1]
x2 = sol.y[2]
y2 = sol.y[3]
x3 = sol.y[4]
y3 = sol.y[5]
```

```python
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)

```

# Langrangian Points and Stability



Look at sources for link to image.

Lagrangian points are a special set of points that provide equilibrium in the three body problem for a small object interacting with two bodies that are more massive than it. The conditions for such a special stable case are: $m_{1}$ >>$m_{2}$ >>$m_{3}$.  The $L_{4}$ and $L_{5}$ points are stable provided that the mass of the primary body (e.g. the Earth) is at least 25 times the mass of the secondary body , and the mass of the secondary is at least 10 times that of the tertiary (wikipedia Lagrangian points).

For this next part of the code, we will look at points $L_{4}$ and $L_{5}$, let us try verifying that $m_{1}$/$m_{2}$ > 25 for a stable solution provided that $m_{3}$ is way smaller than both.

### Conclusion of this analysis

In those graphs, I only graphed the x-coordinate of the trajectory and I kept all the other inital conditions the same for simplicity. Notice that in both cases, it is the green path which corresponds to the $x_{3}$ which is the x coordinate of $m_{3}$ (the smallest object) that is the most sensitive to changes in our initial conditions. So, if we only focus on what happens on $x_{3}$ when we change the intial conditions for mass, that could give us a good prediction of what mass ratio of $m_{1}$ and $m_{2}$ gets us to a stable configuration. We will set $m_{2}$ to 1 so that the value of $m_{1}$ we find directly gives us the ratio directly since anything divided by 1 is itself. 

A stable solution is going to be cyclical. What we can do is examine the cyclical behavior of $x_{3}$ with time as we change the ratio of masses. A good way to estimate the cyclical behavior is to try to fit a straight line to the data. Cyclical behavior will give us a slope of almost zero. We will split the data into quarters. We will look at the cyclical behavior of the second and third quarter. Some quarters might deviate a little bit from cyclical behavior and have a slope that is a little off than zero. To fix that we can multuply the  linear slope of the second and third quarter, if the product is still close to zero than we know that we have approximate cyclical behavior on average and we have a stable solution.

Note that here I am not trying to use the "curve_fit" function to fit a specific function to the data, but instead just as a tool to check if the data is on average cyclical and therefore more in equilibrium.

Let's try implementing that:
